```python
# Run Code_005
import numpy as np
import matplotlib.pyplot as plt

# Define MSE function
def mse(y, y_pred):
    return np.mean((y - y_pred) ** 2)

# Generate data for y and y_pred
y = 1  # Actual values (constant)
y_pred = np.linspace(0, 2, 400)  # Predicted values (varying)

# Compute MSE values
mse_values = (y - y_pred) ** 2  # MSE for each predicted value

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(y_pred, mse_values, label=r'MSE$(\hat{y}, y) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$', color='green', lw=2)

# Add details to the graph
plt.title("Mean Squared Error (MSE) Plot", fontsize=14)
plt.xlabel(r'$\hat{y}$ (Predicted Values)', fontsize=12)
plt.ylabel(r'MSE$(\hat{y}, y)$', fontsize=12)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(1, color='r', linestyle='--', label=r'$y=1$')  # Line where y=1
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='best')

# Display the plot
plt.show()
```
