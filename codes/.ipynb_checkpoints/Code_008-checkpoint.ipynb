```python
# Run Code_008
from graphviz import Digraph
from IPython.display import Image, display

def plot_neural_network():
    dot = Digraph(comment='Neural Network', format='png')

    # Set global attributes for the graph
    dot.attr(dpi='300', rankdir='LR', size='4,2')
    dot.node_attr['style'] = 'filled'
    dot.node_attr['fillcolor'] = '#e6f7ff'

    # Input Layer (Column 1)
    dot.node('X1', 'X1 (Input)', shape='circle', width='0.7', fillcolor='#ffcc99')
    dot.node('X2', 'X2 (Input)', shape='circle', width='0.7', fillcolor='#ffcc99')
    dot.node('X3', 'X3 (Input)', shape='circle', width='0.7', fillcolor='#ffcc99')

    # Hidden Layer (Column 2) - Reduce the size of the circles
    dot.node('H1', 'H1 (Hidden Neuron)', shape='circle', width='0.9', fillcolor='#c2f0c2')
    dot.node('H2', 'H2 (Hidden Neuron)', shape='circle', width='0.9', fillcolor='#c2f0c2')

    # Output Layer (Column 3)
    dot.node('O', 'O (Output)', shape='circle', width='0.7', fillcolor='#ffb3e6')

    # Edges: Input Layer -> Hidden Layer (with weights)
    dot.edge('X1', 'H1', label='W1', color='#6666ff')
    dot.edge('X1', 'H2', label='W2', color='#6666ff')
    dot.edge('X2', 'H1', label='W3', color='#6666ff')
    dot.edge('X2', 'H2', label='W4', color='#6666ff')
    dot.edge('X3', 'H1', label='W5', color='#6666ff')
    dot.edge('X3', 'H2', label='W6', color='#6666ff')

    # Edges: Hidden Layer -> Output Layer (with weights)
    dot.edge('H1', 'O', label='W7', color='#ff6666')
    dot.edge('H2', 'O', label='W8', color='#ff6666')

    # Adding Biases (represented as separate nodes connected to each neuron)
    dot.node('B1', 'Bias (b)', shape='circle', style='dotted', fillcolor='#ffff99', width='0.6')
    dot.edge('B1', 'H1', color='#ff9933')
    dot.edge('B1', 'H2', color='#ff9933')
    dot.edge('B1', 'O', color='#ff9933')

    # Add details to the graph with descriptions
    dot.attr(labeljust='c', labelloc='t', label='Forward Propagation')

    # Render the graph and display it within Jupyter notebook
    file_path = dot.render('/tmp/neural_network_column_colored', cleanup=True)
    display(Image(filename=file_path))

# Call the function to plot the neural network
plot_neural_network()
```
