```python
# Run Code_006
import numpy as np
import matplotlib.pyplot as plt

# Define MSE function
def mse(y, y_pred):
    return np.mean((y - y_pred) ** 2)

# Generate 100 actual values (random between 0 and 2)
np.random.seed(42)  # For reproducibility
y = np.random.uniform(0, 2, 100)

# Generate 100 predicted values (random between 0 and 2)
y_pred = np.random.uniform(0, 2, 100)

# Compute MSE values
mse_value = mse(y, y_pred)

# Scatter plot of actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, color='blue', alpha=0.6, label='Actual vs Predicted')
plt.plot([0, 2], [0, 2], 'r--', label='Perfect Prediction Line')  # y = x line

# Add details to the graph
plt.title(f"Mean Squared Error (MSE): {mse_value:.4f}", fontsize=14)
plt.xlabel(r'Actual Values ($y$)', fontsize=12)
plt.ylabel(r'Predicted Values ($\hat{y}$)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
