```python
#Run Code_007
import numpy as np
import matplotlib.pyplot as plt

# Define Cross-Entropy Loss function
def cross_entropy_loss(y, y_pred):
    return -np.sum(y * np.log(y_pred + 1e-9))  # Adding small value to prevent log(0)

# Define Hinge Loss function
def hinge_loss(y, y_pred):
    return np.mean(np.maximum(0, 1 - y * y_pred))

# Generate data for classification (Binary case: y  {0,1}, y_pred  [0,1])
y = np.array([1] * 100)  # True labels (all ones for simplicity)
y_pred = np.linspace(0.01, 0.99, 100)  # Predicted probabilities (avoiding 0 for log)

# Compute loss values
cross_entropy_values = [-cross_entropy_loss(np.array([1, 0]), np.array([p, 1 - p])) for p in y_pred]

# Generate data for Hinge Loss (y  {-1,1}, y_pred is continuous)
y_hinge = np.array([1] * 100)  # True labels for Hinge Loss (+1 class)
y_pred_hinge = np.linspace(-2, 2, 100)  # Predicted scores (before activation)

# Compute Hinge Loss
hinge_loss_values = [hinge_loss(y_hinge, np.array([p])) for p in y_pred_hinge]

# Plot Cross-Entropy Loss
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(y_pred, cross_entropy_values, color='blue', lw=2, label="Cross-Entropy Loss")
plt.xlabel("Predicted Probability ()")
plt.ylabel("Loss")
plt.title("Cross-Entropy Loss Curve")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

# Plot Hinge Loss
plt.subplot(1, 2, 2)
plt.plot(y_pred_hinge, hinge_loss_values, color='red', lw=2, label="Hinge Loss")
plt.xlabel("Predicted Score ()")
plt.ylabel("Loss")
plt.title("Hinge Loss Curve")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

# Show plots
plt.tight_layout()
plt.show()
```
